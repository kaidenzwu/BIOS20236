# Dimensionality reduction through Principal Component Analysis

Milena is a microbiologist who is sampling bacterial communities in the soil. She is able to measure the abundances of 100 different species in each given sample ($\bar x_i \in \mathbb{R}^{100} $) and hopes to use this data to learn something about how the composition of the community varies across these samples.


\begin{equation}
X = \
\begin{bmatrix}
\begin{bmatrix}
    x_1^{(1)} ,
    x_1^{(2)} , \ldots
    x_1^{(100)}
\end{bmatrix}, \\
\begin{bmatrix}
    x_2^{(1)} ,
    x_2^{(2)} , \ldots
    x_2^{(100)}
\end{bmatrix} , \\ \vdots \\
\begin{bmatrix}
    x_N^{(1)} ,
    x_N^{(2)} ,  \ldots
    x_N^{(100)}
\end{bmatrix}
\end{bmatrix}
\end{equation}

To clear up some notation, $X$ is a matrix of shape ($N_{samples}$, 100) with each row is denoted as $\bar x_i$. The subscript denotes sample number while the superscript denotes the abudnance of a particular species. Each $\bar x_i$ has 100 entires which correspond to a single measurement of the abundances of 100 microbes.

What Milena does not know is that the community composition is fully characterized by 3 major environmental parameters: $\text{pH}$, $NO_3^-$, and $c_b$. To put some names to these parameters, we can say that $\text{pH}$ is the pH, $NO_3^-$ is the abundance of nitrate, and $c_b$ is the community biomass of a particular soil sample $x_i$. Unfortunately for her, **her experimental measurements $x_i$ did not directly quantify $\text{pH}$, $NO_3^-$, and $c_b$**, but her recordings are linearly related to these hidden variables.  Her goal is to find out how these hidden variables affect abundances.


$$\bar x_i = [\text{pH}]_i \bar v_{1} + [NO_3^-]_i \bar v_{2} + [c_b]_i \bar v_{3}$$

Please note that **$\bar x$ and $\bar v$ are 100 dimensional vectors**. On the other hand, $[\text{pH}]_i$, $[NO_3^-]_i$, and $[c_b]_i$  are the pH, nitrate, and community biomass of sample $x_i$ which **are numbers ($d=1$)**. Each $v$ vector is constant and of unit length (the magnitude is equal to 1). You can think of each $\bar v$ vector as how much that its corresponding parameter effects the community of microbes in the sample and represent the **major axes of variation**. The way to think about this is that changes in pH alter the composition of the community in a manner described by the vector $ \bar v_{1}$. $ \bar v_{1}$ is a *single* vector in the 100 dimensional space of microbial abundances in her samples.

If we think of $\bar x$, $\text{pH}, NO_3^-,$ and $c_b$ as **deviations from the mean**, we can effectively ignore the absolute scale of these variables and think of each of these parameters are drawn from Gaussians centered at 0, but with different variances.

$$[\text{pH}]_i \sim \mathcal{N}(\mu = 0, \sigma^2 = 20)$$
$$[NO_3^-]_i \sim \mathcal{N}(\mu = 0, \sigma^2 = 5)$$
$$[c_b]_i \sim \mathcal{N}(\mu = 0, \sigma^2 = 0.5)$$

Even though each sample has 100 dimensions, we know that the entire dataset is built on just 3 major axes of variation ($\bar v_1$, $\bar v_2$, and $\bar v_3$ ) that were not directly measured (poor Milena). You can see here that pH has the largest $\sigma^2$ and thus, **explains the most variances** in Milena's samples.  What this means is that pH varies the most, and thus the biggest changes in community composition happen along the direction defined by $ \bar v_{1}$.  Similar argument for $ \bar v_{2}$ and $ \bar v_{3}$.

```python
import numpy as np
from numpy import linalg as LA
```

```python
#Your code here!
v1 = []
v2 = []
v3 = []
samples = np.arange(1,501)
X = []
peeH = []
Noo3 = []
ceebee = []

for i in range(100):
    v1rand = np.random.normal()
    v1.append(v1rand)
    v2rand = np.random.normal()
    v2.append(v2rand)
    v3rand = np.random.normal()
    v3.append(v3rand)

v1 = v1 / LA.norm(v1)
v2 = v2 / LA.norm(v2)
v3 = v3 / LA.norm(v3)

print (str(LA.norm(v1)) + ', ' + str(LA.norm(v2)) + ', ' + str(LA.norm(v3)))

#Time to find angle

theta12 = np.arccos(np.dot(v1,v2))
theta23 = np.arccos(np.dot(v2,v3))
theta13 = np.arccos(np.dot(v1,v3))

print('angles')
print(theta12*(180/np.pi))
print(theta23*(180/np.pi))
print(theta13*(180/np.pi))
print('pi/2 : ' + str(np.pi / 2))

for i in samples:
    pH = np.random.normal(0, 4.47213595)
    NO3 = np.random.normal(0, 2.23606798)
    cb = np.random.normal(0, 0.707106781)
    X.append((pH * v1) + (NO3 * v2) + (cb * v3))
    peeH.append(pH)
    Noo3.append(NO3)
    ceebee.append(cb)

print(np.shape(X))
for i in range(0,500):
    X[i] = X[i] + np.random.normal(0, 1, (100))

print(np.shape(X))
#np.reshape(X, (500, 100))
#print(X[0])
X = np.array(X)
np.var(X)
```

3. These dot products are 0.9893515315681988, -0.9787087379585614, -0.3195829765125616. The reason the dot products should be either 1 or -1 is that the eigenvectors generated are supposed to capture the most variance in order, which means they should be aligned in direction to the vectors v1, v2, v3 that we used to introduce variance. The dot product formula is v1 dot pc 1 = ||v1|| * ||pc1|| cos(theta), and since both v1 and pc1 are normalized, the dot product is soley dependent on the angle. If the angles are pointing in the same or exactly opposite directions, the result would be 1 or -1. The dot product can be -1 as well, suggesting a negative correlation.
4. The PCA recovered the first two axes very well, but it had a lot of trouble with identifying the 3rd axis, which would normally correlate with v3. This is because the 3rd axis, which corresponds to the variance added by community biomass, is drowned out by the noise.
