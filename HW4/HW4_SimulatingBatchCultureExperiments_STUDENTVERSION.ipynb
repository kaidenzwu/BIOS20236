# HW2: simulating ODEs to investigate coexistence in communities

(55 pts total) First we will setup a simulation framework for doing serial batch culture that we discussed in lecture.  The situation we are simulating here is a community of species that is growing on a bunch of substitable resources.  An example would be many strains (e.g. those from Project 2) growing on many different carbon sources.

The Model:

$$ \dot{N}_i = N_i \sum_j \frac{r_{i,j}\gamma_{i,j} R_j}{K_{i,j} + R_j}$$
$$ \dot{R_j} = - N_i \sum_j \frac{r_{i,j} R_j}{K_{i,j} + R_j}$$

The index $i$ is over species, the index $j$ is over resources (e.g. carbon sources).  Note the sum in the first equation indicates that these resources are substitutable.  The resources ($R$) and abundances ($N$) need to have some initial values at the beginning of the simulation.  We will ignore the $K_{i,j}$ in what follows and set them all to a small constant, an approximation that is biologically motivated -- these numbers tend to be small.  The parameters are as follows.

The $r_{i,j}$ are the resource uptake rates for species $i$ on resource $j$.
The $\gamma_{i,j}$ are the growth yields for strain $i$ on resource $j$.

**Question 1:** (2 pts) given the equations above and the fact that $N_i$ has units of biomass and $R_j$ has units of concentration (mM) what are the units of the $r_{i,j}$ and $\gamma_{i,j}$?

**Question 2:** (2 pts) the product $r_{i,j} * \gamma_{i,j}$ has what units?  What does this product correspond to biologically?

1. $r_{i,j}$ has units of concentration (mM) / biomass(g) * time(s), and $\gamma_{i,j}$ has units of biomass * inverse concentration, or biomass(g) / concentration (mM)
2. This product has overall units of 1 / time, and this product corresponds to a the percent increase over time.

# Step 2: running a serial dilution simulation of one strain growing on one resource.

Below is a set of functions that allow you to integrate the ODEs above using ODEINT that we discussed in lecture.  LOOK at this code and make sense of it, then answer the fllowing questions.

**Question 4** (6 pts) What terms in the equations above are 'species_abs' and 'resource_concs' describing?  What is the function 'F' doing with these two functions?  What does 'CRM' do?

4. species_abs returns the N * (r * gamme * R / R + K) term of the dN/dt equation or the overall instantaneous rate of change for biomass, while resource_concs describes the R * (r * R) / K + R) portion of the dR/dt equation, which represents the instananeous rate of change of resource concentration over time. F is a function that defines a matrix of dN/dt and dR/dt for all the species values and all the resource values. CRM takes all these instantaneous rates of change from F and integrates them over a set range of times. 

```python
import numpy as np
from scipy.integrate import odeint

############################
#### primary batch function ######
############################
def CRM(y0,t,p,n_species,n_resources):
    #sol = odeint(F, y0, t, Dfun=J, args=(p,n), rtol=1e-6)
    sol = odeint(F, y0, t, args=(p,n_species,n_resources))#, rtol=1e-6)
    return sol

###########################
#### batch helper functions #####
###########################
def F(y,t,p,n_species,n_resources):
    #ODE RHS
    Fout = np.zeros(n_species+n_resources)
    for i in range(n_species):
        for j in range(n_resources):
            Fout[i]  = Fout[i]+species_abs(p, y, i, j, n_species)
    for j in range(n_resources):
        for i in range(n_species):
            Fout[n_species+j] = Fout[n_species+j]+resource_concs(p, y, i, j, n_species)
    return Fout

def species_abs(p,y,species_idx,r_idx,n_species):
    r     = p[species_idx, r_idx, 0]
    gam   = p[species_idx, r_idx, 1]
    K     = p[species_idx, r_idx, 2]
    x     = y[species_idx]
    R     = y[n_species+r_idx]
    return gam*r*x*R/(K + R) 

def resource_concs(p,y,species_idx,r_idx,n_species):
    r     = p[species_idx, r_idx, 0]
    gam   = p[species_idx, r_idx, 1]
    K     = p[species_idx, r_idx, 2]
    x     = y[species_idx]
    R     = y[n_species+r_idx]
    return -1*r*x*R/(K + R) 
```

**Example code** Here is a simulation of a single species (index: 0,0 from the uptakerate/yield matrices uploaded above)

```python
# 48h growth period / cycle -- this is the time vector that will be integrated.
t = np.logspace(-2, 2, 1080)
t = [i for i in t if i <= 48]

#select just the first strain and it's growth on just one resource (arabinose)
df_UR = np.array(df_uptake.iloc[0,0]).reshape(-1,1) # this selects the uptake rate for the first strain, first resource
df_Y = np.array(df_yields.iloc[0,0]).reshape(-1,1) # does the same for the yields

r_mM = 5 # starting concentration of the yields
n_s,n_r = [1,1] # defines the number of species and number of resources
params_vals = np.zeros((n_s, n_r, 3)) # defines the parameter values matrix that is passed to CRM above
print(params_vals)
params_vals[:,:,0] = df_UR # uptake rates
print(params_vals)
params_vals[:,:,1] = df_Y # gammas (yields)
print(params_vals)
params_vals[:,:,2] = 0.01 # affinities -- again just fixed for every strain/resource combination
print(params_vals)

y0 = np.zeros(n_s + n_r) # note in the code we pass strain abundances and resources as one apended vector called y.
y0[:n_s] = 0.01 # species starting amount [OD]
y0[n_s:] = r_mM # resources starting amount [mM]

# at this point all the variables and parameters are initialized!!  The next cell runs the simulation.
```

```python
num_dilutions = 4 # number of dilutions
dilution_factor = 100 # dilution factor between dilutions
R0 = r_mM; # this is the concentration of the resources after each dilution.
all_dynamics = np.zeros((len(t),n_s+n_r,num_dilutions)) # a matrix to save the data.

# each iteration of this loop is a dilution cycle
for cycle in range(1,num_dilutions+1):
    print (f'On cycle {cycle}')
    dynamics = CRM(y0,t,params_vals,n_s,n_r) #simulate one cycle.
    all_dynamics[:, :,cycle-1] = dynamics
    
    # check for extinctions (defined as the abundance of a strain going below a threshold).  If something goes extinct it cannot grow again in the next cycle (duh)
    for x in range(n_s):
        if y0[x] > 1e-6:
            y0[x] = dynamics[-1,x] / dilution_factor
            print('chillin')
        else:
            y0[x] = 0
            print('yikes')
        y0[n_s:] = R0 #+ dynamics[-1,n_s_picked:] / dilution_factor
```
